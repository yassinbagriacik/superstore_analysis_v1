```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")

print(df.head())
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

subcat_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
top5_subcat = subcat_sales.head()
print(top5_subcat)

plt.figure(figsize = (8,5))
top5_subcat.plot(kind ='bar', color='skyblue')
plt.title('En Çok Satış Yapılan 5 Alt Kategori')
plt.xlabel('Alt Kategori')
plt.ylabel('Toplam Satış')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
