```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('train.csv') # dosyayı içe aktarır veya okuma

print(df.columns)

print(df.head()) # ilk 5 satırı görüntüle

print(df['City'].value_counts().head()) # en çok satış yapan ilk 5 şehir

print(df.groupby('Category')['Sales'].sum()) # kategoriye göre toplam satış

# Aylık satış grafiği

df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df.set_index('Order Date')['Sales'].resample('M').sum().plot(figsize=(10,4))
plt.title('Aylık Satış Grafiği')
plt.xlabel('Tarih')
plt.ylabel('Toplam Satış')
plt.show()


```
